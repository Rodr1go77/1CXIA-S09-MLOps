```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline

print("Bibliotecas carregadas com sucesso!")
```

## 📊 Análise Exploratória de Dados (EDA)
Gerando dados fakes para validar a distribuição de sentimentos da base histórica.

```python
data = {'sentimento': ['POSITIVO', 'NEGATIVO', 'POSITIVO', 'POSITIVO', 'NEGATIVO', 'POSITIVO', 'POSITIVO']}
df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
sns.countplot(x='sentimento', data=df, palette='viridis')
plt.title("Distribuição de Sentimentos (Base de Teste)")
plt.show()
```

## 🚀 Carregamento do Modelo Pré-treinado
Usando o pipeline do Hugging Face para carregar o DistilBERT.

```python
# FIX: Se der erro de memória, tentar modelo menor
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

def predizer_sentimento_v1(texto):
    print(f"Analisando o texto: {texto}")
    resultado = classifier(texto)
    return resultado

print("Modelo carregado!")
```

```python
# TESTE 01
frase = "O atendimento no caixa eletrônico foi muito rápido hoje!"
res = predizer_sentimento_v1(frase)
print(f"Resultado: {res}")
```
